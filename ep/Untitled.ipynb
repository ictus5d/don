```python
import numpy as np
import matplotlib.pyplot as plt

N = 20
theta = np.linspace(0.0, 2*np.pi, N,endpoint=False)
radii = 10 * np.random.rand(N)
width = np.pi / 4 * np.random.rand(N)

ax = plt.subplot(111, polar = True)
bars = ax.bar(theta, radii, width=width, bottom=0.0)

#use custom colors and opacity
for r, bar in zip(radii, bars):
    bar.set_facecolor(plt.cm.jet(r / 10.))
    bar.set_alpha(0.5)

plt.show()
```

```python
def generate_data(n,generator_type):
    e_values = []
    for i in range(n):
        if generator_type == 'U':
            e = np.random.uniform(0,1)
        else:
            e = np.random.randn()
        e_values.append(e)
    return e_values

data = generate_data(100, 'U')
plt.plot(data)
plt.show()
```

```python
#exercice 1
def factorial(n):
    result = 1
    for i in range(n):
        result = result * (i+1)
    return result
```

```python
#exercice 2

import matplotlib.pyplot as plt

from numpy.random import uniform
def binomial_rv(nbr_essai,probability):
    nbr_cible = 0
    for i in range(nbr_essai):
        U = uniform()
        if U < probability:
            nbr_cible += 1         
    return nbr_cible
```

```python
%matplotlib inline
```

```python
binomial_rv(100,0.5)
```

```python
binomial_rv(200,0.5)
```

```python
#exercice 3 methode 1

import numpy as np

n = 100000

count = 0
for i in range(n):
    u,v = np.random.uniform(), np.random.uniform()
    d = np.sqrt((u-0.5)**2 + (v-0.5)**2)
    if d < 0.5:
        count += 1
        
area_estimate = count/n

print(area_estimate * 4)
```

```python
#exercice 3 mthode 2

import numpy as np

nbr_tir = 100000
nbr_cible = 0

for i in range(n):
    x,y  = np.random.uniform(), np.random.uniform()
    d = np.sqrt((x-0.5)**2 + (y-0.5)**2)
    if d <= 0.5:
        nbr_cible += 1

rapport = nbr_cible/nbr_tir

print(rapport * 4)
```

```python
#exercice 3, mthode 3

import numpy as np

nbr_tir = 100000
nbr_cible = 0

for i in range(n):
    x,y  = np.random.uniform(), np.random.uniform()
    d = np.sqrt(x**2 + y**2)
    if d <= 1:
        nbr_cible += 1

rapport = nbr_cible/nbr_tir

print(rapport * 4)
```

```python
#exercice 6

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def plot_standard_normal(variance):
    length = 200
    current_x = 0
    x_values = []
    
    for i in range(length):
        x_values.append(current_x)
        current_x = variance*current_x + np.random.randn()
    return x_values

n = 3
v = [0.0, 0.8, 0.98]
colors = ['blue','orange','green']
for i in range(3):
    value = plot_standard_normal(v[i])
    plt.plot(value, c=colors[i], label='variance='+ str(v[i]))
plt.legend()
plt.show()
```
