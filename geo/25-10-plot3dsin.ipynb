```python
def f(x,y):
     return cos(0.001+y^2+x^2)/sqrt(x^2+y^2+0.001)
P = plot3d(f,(-3,3),(-3,3), adaptive=True, 
           color=rainbow(80, 'rgbtuple'), 
           max_bend=0.01, max_depth=1)
P.show(fontsize=6)
```

```python
def f(x,y):
     return math.cos(y^2+x^2)/math.sqrt(x^2+y^2+0.1)
P = plot3d(f,(-1,1),(-1,1), adaptive=True, 
           color=rainbow(50, 'rgbtuple'), 
           max_bend=0.1, max_depth=1)
P.show(fontsize=6)
```
