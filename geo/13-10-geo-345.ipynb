```python
%display latex
g= Graphics() 
g+= line([(1,2),(4,6),(-4,12),(1,2)],color="green") 
g+= text("P",(0.2,2.5),fontsize=6,color="black") 
g+= text("Q",(3.8,7),fontsize=6,color="blue") 
g+= text("R",(-4,11),fontsize=6,color="red") 
show(g,figsize=2,fontsize=4,frame=True,
  gridlines="minor",gridlinesstyle=dict(color="orange",
                         linestyle="-",alpha=0.4)  )
```
