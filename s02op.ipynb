```python
import numpy as np               
import matplotlib.pyplot as plt 
%matplotlib inline
```

##### (Pas de changement pour convertir python en sage).

Le profil de l'onde à propager: une gaussienne multipliée par un cosinus.

Le nombre de points d'échantillonnage du graphe (nb_points = 1000)

Le nombre de courbes à représenter (nb_courbes =3)

```python
def f(u,k=10):
    return np.exp(-3*u**2) * np.cos(k*u-5)
nb_points  = 1000   
nb_courbes = 3
```

##### Tout d'abord la visualisation spatiale 

$x=np.linspace(-2,2,$ nb_points$)$  Echantillonnage en position

$t=np.linspace(0,5,$nb_courbes$)$  On regarde le profil à différents temps

$c=0.2$  Vitesse de propagation de l'onde

```python
x = np.linspace(-2,2,nb_points)   
t = np.linspace(0,5,nb_courbes)  
c = 0.2
```

$for $ $ti$  $in$ $t:$
     $fi = f(x-c*ti)$  (Echantillonnage du profil pour les différents x)
     $plt.plot(x,fi,label='$t={}$'.format(round(ti,1)))$   (Affichage)


```python
for ti in t:
    fi = f(x-c*ti)
    plt.plot(x,fi,label='$t={}$'.format(round(ti,1)))
```

##### La cosmétique


```python
plt.title('Profil spatial pour differents temps')
plt.xlabel('Position $x$')
plt.ylabel("Profil de l'onde")
plt.legend()
#plt.savefig('S02_onde_progressive_spatial.png')
#plt.clf() ;
```

##### Tout d'abord la visualisation spatiale

Echantillonnage en temps

On regarde le profil à différentes positions

Vitesse de propagation de l'onde

```python
t = np.linspace(0,10,nb_points)  
x = np.linspace(0,0.6,nb_courbes)  
c = 0.2
```

Echantillonnage du profil pour les différents $t$

Affichage et cosmétique

```python
for xi in x:
    fi = f(xi-c*t) 
    plt.plot(t,fi,label='$x={}$'.format(round(xi,1))) 
plt.title("Profil temporel pour differentes positions")
plt.xlabel('Temps $t$')
plt.ylabel("Profil de l'onde")
plt.legend()
plt.show()
#plt.savefig('PNG/S02op.png');
```
